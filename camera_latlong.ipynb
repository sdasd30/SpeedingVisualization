```python
import time
import pandas as pd
import numpy as np
from dask import dataframe as df1
from geopy.geocoders import Nominatim
```

```python
# initialize Nominatim API 
geolocator = Nominatim(user_agent="geoapiExercises")

#Create compression algorithm
compression_opts = dict(method='zip',
                        archive_name='out.csv')  
```

```python
start = time.time()
size_of_data = 0
df = df1.read_csv("./data/speeding_violations.csv", dtype={'current_amount_due': 'float64',
       'hearing_disposition': 'object',
       'zipcode': 'object',
        'violation_code': 'object'}, low_memory = False)
print("Load completed in " + str(time.time() - start) + "seconds")
df.head(10)
```

```python
len(df)
pdf = df.compute()
```

```python
streets = pdf.groupby('violation_location').violation_location.count()
print(streets)
```

```python
streetnames = streets.index.tolist()
```

```python
print(streetnames)
```

```python
testStreet = "1754 N PULASKI RD"
```

```python
fdf = pdf[pdf['violation_location'] == testStreet] #fDF for Filtered DF
#print(len(fdf))
ydf = fdf[fdf['year'] == 2018] #yDF for year DF
mdf = ydf[ydf['month'] == 1]
len(mdf)
```

```python
output = pd.DataFrame(columns=['Camera','Year','Month','Total Repeat Offenders',
                               'Total New Offenders', 'New Offenders This Year',
                               'Repeat Offenders This Year', 'New Offenders This Month',
                              'Repeat Offenders This Month'])
dataList = []

amount = len(streetnames)
cnt = 0
testset = streetnames#[0:3]
#testset = ["1754 N PULASKI RD", "7738 S WESTERN"]
for street in testset:
    cnt += 1
    print("Street",street,"complete ["+str(cnt) + "/" + str(amount) + "]")
    fdf = pdf[pdf['violation_location'] == street]
    
    totEnc = {} #total encounters
    totalOffenders = 0
    totalUnique = 0
    
    
    started = False
    startyear = 0
    startmonth = 0
    endyear = 0
    endmonth = 0
    
    for year in range(2003, 2019):
        ydf = fdf[fdf['year'] == year] #yDF for year DF
        
        if (len(ydf) >0):
            for month in range(1,13):
                if (month == 6 and year == 2018):
                    break
                monthNew = 0
                monthRepeat = 0
                halfRepeat = 0
                yearRepeat = 0
                lifeRepeat = 0
                
                monthEncountered = {}
                
                mdf = ydf[ydf['month'] == month]
                
                if (len(mdf) > 0):
                    endyear = year
                    endmonth = month
                    if (not started):
                        timestamp = -1
                        started = True
                        startyear = year
                        startmonth = month
                    timestamp += 1
                    
                        
                    binned = mdf.groupby('license_plate_number').license_plate_number.count()
                    
                    for plate,count in binned.items():
                        totalOffenders += count
                        if (plate in totEnc):
                            lifeRepeat += count
                            check = timestamp - totEnc[plate] 
                            #print(check)
                            if (check == 0):
                                monthRepeat += count
                            if (check <= 12):
                                yearRepeat += count
                            if (check <= 6):
                                halfRepeat += count
                            totEnc[plate] = timestamp
                        else:
                            totEnc[plate] = timestamp
                            monthNew += 1
                            totalUnique += 1
                            lifeRepeat += (count - 1)
                            monthRepeat += (count - 1)
                            yearRepeat += (count - 1)
                            halfRepeat += (count - 1)
                            
                            
                                                
                if (started):
                    dataList.append({"Camera":street,
                                    "Timestamp":timestamp,
                                    "Year": year,
                                    "Month": month,
                                    "Total Offenses":totalOffenders,                                     
                                    "Total Unique Offenders":totalUnique,
                                    "New Offenders This Month":monthNew,
                                    "Repeat Offenders":lifeRepeat,
                                    "Repeat Offenders Within The Year":yearRepeat,
                                    "Repeat Offenders Within Past 6 Months":halfRepeat,
                                    "Repeat Offenders Within The Month":monthRepeat})
                
    
    #print("Ends:",endyear,endmonth)
    #if (endyear != 2018 or endmonth != 12):
    #    print("Street:",street, "is an exception", endyear,endmonth)
        
    finalFrame = pd.DataFrame.from_dict(dataList)
    #print(finalFrame.head())
    #print("Street:",street,"Starts:",startyear,startmonth,"Ends:",endyear,endmonth, "New:",totalnew,"Repeats:",totalrepeat)
```
