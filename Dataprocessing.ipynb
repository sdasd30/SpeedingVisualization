```python
import time
import pandas as pd
import numpy as np
from dask import dataframe as df1
from geopy.geocoders import Nominatim
```

```python
# initialize Nominatim API 
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
start = time.time()
size_of_data = 0
df = df1.read_csv("./data/chicago_camera_tickets.csv", dtype={'current_amount_due': 'float64',
       'hearing_disposition': 'object',
       'zipcode': 'object',
        'violation_code': 'object'}, low_memory = False)
print("Load completed in " + str(time.time() - start) + "seconds")
df.head(10)
```

```python
start = time.time()
print("When grouped by violation, the tickets would be organized as")
df = df[(df.violation_description == "SPEED VIOLATION 11+") | (df.violation_description == "SPEED VIOLATION 6-10")]
df = df.compute()
print("Computation completed in " + str(time.time() - start) + " seconds")
df.head()
```

```python
zipcodes = {}
unknown = []
i = 0
for street, count in streets.items():
    i += 1
    try:
        time.sleep(1)
        place = "Chicago, " + street
        location = geolocator.geocode(place)
        if (location != None):
            data = location.raw
            loc_data = data['display_name'].split()
            zipcodes[street] = loc_data[-3]
            print("New Zipcode: ", loc_data[-3])
            print(i, " complete")
            i+=1
        else:
            unknown.append(street)
    except:
        print("Connection refused, waiting for 5 seconds")
        time.sleep(5)
        try:
            place = "Chicago, " + street
            location = geolocator.geocode(place)
            if (location != None):
                data = location.raw
                loc_data = data['display_name'].split()
                zipcodes[street] = loc_data[-3]
                print("New Zipcode: ", loc_data[-3])
                print(i, " complete")
                i+=1
            else:
                unknown.append(street)
        except:
            print("Second connection failed, sleeping 5 seconds and skipping address")
            unknown.append(street)
            time.sleep(5)
            
    
print(zipcodes)
    
```
